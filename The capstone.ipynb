```python
x=df.drop('Class',axis=1)
y=df['Class']
print('Shape of x',x.shape)
print('Shape of y',y.shape)
```

```python
#now seperate fraud and nonfraud data.
Fraud_d=df[df['Class']==1]
Nonfraud_d=df[df['Class']==0]
print('Shape of Fraud data',Fraud_d.shape)
print('Shape NonFraud data',Nonfraud_d.shape)
```

```python
#now there are two ways to go from here as data is highly unbalanced
#we need a method to tackle this problem.
#we can undersample the data which will tone down the majority level data
#to the minority level with ratio 1:1.
#we can oversample using SMOTE(Synthetic Minority Oversampling Technique) which 
#is technique used to artificially create minority data points and upscalling them
#to the level of majority data.
```

```python
from imblearn.under_sampling import NearMiss
```

```python
sam=NearMiss()
x_un,y_un=sam.fit_resample(x,y)
print('Shape of Fraud case undersampled',x_un.shape)
print('Shape of NonFraud case undersampled',y_un.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_un,y_un,test_size=0.3,random_state=42)
```

```python
#now lets start using the classifiers with the undersampled data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,precision_recall_curve,precision_score,recall_score,classification_report,accuracy_score,f1_score
logr=LogisticRegression()
logr.fit(x_train,y_train)
y_prdt=logr.predict(x_test)
cnf_mat=confusion_matrix(y_test,y_prdt)
sn.heatmap(pd.DataFrame(cnf_mat),annot=True,cmap="YlGnBu",fmt='g')
pt.ylabel('TRUE LABEL')
pt.xlabel('PREDICTED LABEL')
labels=('NON_FRAUD','FRAUD')
logpre=precision_score(y_test,y_prdt,)
logrec=recall_score(y_test,y_prdt,)
logf1=accuracy_score(y_test,y_prdt,)
logf2=accuracy_score(y_test,y_prdt,)
print(classification_report(y_test,y_prdt,target_names=labels))
```

```python
#accuracy of this model is good as there seems to be no bias
#now lets try some more classifier to find out which is best
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_prdtd=dtc.predict(x_test)
cnf_mat=confusion_matrix(y_test,y_prdtd)
sn.heatmap(pd.DataFrame(cnf_mat),annot=True,cmap="YlGnBu",fmt='g')
pt.ylabel('TRUE LABEL')
pt.xlabel('PREDICTED LABEL')
labels=('NON_FRAUD','FRAUD')
decpre=precision_score(y_test,y_prdtd,)
decrec=recall_score(y_test,y_prdtd,)
decf1=accuracy_score(y_test,y_prdtd,)
decf2=f1_score(y_test,y_prdtd,)

print(classification_report(y_test,y_prdtd,target_names=labels))
```

```python
from sklearn import ensemble
from xgboost import  XGBClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix,precision_recall_curve,precision_score,recall_score,classification_report,accuracy_score
```

```python
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
yxgb_prdt=xgb.predict(x_test)
xgbpre=precision_score(y_test,yxgb_prdt,)
xgbrec=recall_score(y_test,yxgb_prdt,)
xgbf1=accuracy_score(y_test,yxgb_prdt,)
xgbf2=f1_score(y_test,yxgb_prdt,)

print(classification_report(y_test,yxgb_prdt,target_names=labels))
```

```python
cat=CatBoostClassifier()
cat.fit(x_train,y_train)
ycat_prdt=cat.predict(x_test)
catpre=precision_score(y_test,ycat_prdt,)
catrec=recall_score(y_test,ycat_prdt,)
catf1=accuracy_score(y_test,ycat_prdt,)
catf2=f1_score(y_test,ycat_prdt,)
print(classification_report(y_test,ycat_prdt,target_names=labels))
```

```python
ada=ensemble.AdaBoostClassifier()
ada.fit(x_train,y_train)
yada_prdt=ada.predict(x_test)
adapre=precision_score(y_test,yada_prdt,)
adarec=recall_score(y_test,yada_prdt,)
adaf1=accuracy_score(y_test,yada_prdt,)
adaf2=f1_score(y_test,yada_prdt,)
print(classification_report(y_test,yada_prdt,target_names=labels))
```
